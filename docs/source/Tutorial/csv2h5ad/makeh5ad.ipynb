# Download the cpg0016 source4 dataset to the local folder.

<div class="alert note">
<p>

**Note**

This code uses the AWS CLI to copy files from an S3 bucket to a local directory. 

Specifically, the command aws s3 cp copies files from s3://cellpainting-gallery/cpg0016-jump/source_4/workspace/backend/ to /data/pub/cell/cpg0016_source4/backend on your local machine.

The --no-sign-request flag allows you to access public data without signing in. 

The --recursive option ensures that all files and subdirectories are copied, while --exclude "*.sqlite" prevents any files with the .sqlite extension from being downloaded.
 
This effectively allows you to download the cpg0016 dataset to your specified local directory.

</p>
</div>

```python
# !aws s3 cp --no-sign-request s3://cellpainting-gallery/cpg0016-jump/source_4/workspace/backend/ /data/pub/cell/cpg0016_source4/backend --recursive --exclude "*.sqlite"
```

```python
import pandas as pd
import os
import warnings
warnings.filterwarnings("ignore")
def merge_allscsv(folder_path, data):
    for root, _, files in os.walk(folder_path):
        for file in files:
            if file.endswith('.csv'):
                file_path = os.path.join(root, file)
                da = pd.read_csv(file_path)
                data = pd.concat([data, da])
    return data

df = pd.DataFrame()
path = '/data/pub/cell/cpg0016_source4/backend/'
data = merge_allscsv(path, df)
```

```python
plate = pd.read_csv('/data/pub/cell/cpg0016_source4/plate.csv') 
plate=plate[plate['Metadata_Source']=='source_4']
plate
```

```python
data=pd.merge(data, plate[['Metadata_Plate', 'Metadata_Batch', 'Metadata_PlateType']], 
                        on=['Metadata_Plate'], 
                        how='left')
data
```

```python
well = pd.read_csv('/data/pub/cell/cpg0016_source4/well.csv')
well = well[well['Metadata_Source']=='source_4']
well
```

```python
data=pd.merge(data, well[['Metadata_Plate','Metadata_Well','Metadata_JCP2022']], 
                        on=['Metadata_Plate','Metadata_Well',], 
                        how='left')
data
```

```python
import numpy as np
r,c = np.where(data.isna())
features_to_remove = [_ for _ in list(data.columns[list(set(c))]) if not _.startswith('Metadata_')]
print(f'Removed nan features: {features_to_remove}')
data.drop(features_to_remove, axis=1, inplace=True)
data.shape
```

```python
data
```

```python
features_meta =  [_ for _ in list(data.columns) if _.startswith('Metadata_')]
```

```python
import anndata as ad
X=data.iloc[:,9:-3]
obs=pd.DataFrame()
var = pd.DataFrame(index=data.columns[9:-3])
for feature in features_meta:
    obs[feature]=data.loc[:][feature]
row = [n[0:1] for n in data['Metadata_Well'].values]
col = [n[1:] for n in data['Metadata_Well'].values]
obs['row']=row
obs['col']=col

adata=ad.AnnData(X.values,obs=obs,var=var,dtype='float64')
```

```python
adata.obs['Metadata_Plate'] = adata.obs['Metadata_Plate'].astype(str)
adata.obs['Metadata_Well'] = adata.obs['Metadata_Well'].astype(str)
adata.obs['location'] = adata.obs['Metadata_Plate'] + "_" + adata.obs['Metadata_Well']
adata.obs.set_index('location', inplace=True)
```

```python
adata.obs.rename(columns={'Metadata_Batch': 'batch'}, inplace=True)
adata.obs['batch'] = adata.obs['batch'].apply(lambda x: x.split('_')[-1])
```
